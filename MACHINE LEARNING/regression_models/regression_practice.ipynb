MODEL CREATION

```python
df.head()
```

```python
x= df.iloc[:, :5] #data frame
y= df.iloc[:,5] #series
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtest,xtrain,ytest,ytrain = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
#import the model
from sklearn.linear_model import LinearRegression

#Intialize the model
linreg=LinearRegression()

#train the model

linreg.fit(xtrain,ytrain)

#make predictions
yprd=linreg.predict(xtest)
```

```python
yprd
```

```python
ytest
```

#MODEL EVALUATION

```python
#To find m use .coef_
linreg.coef_
```

```python
# sns.distplot(ytest-yprd)



# MAE: 5.221743847040538
# MSE : 36.39949029502399
# RMSE : 6.033199010062902
# Accuracy:0.9611509969233327

```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae=mean_absolute_error(ytest,yprd)    #MEAN ABSOLUTE ERROR
mse=mean_squared_error(ytest,yprd)     #MEAN SQUARED ERROR
r2 = r2_score(ytest,yprd)  #How good is your module
rmse = np.sqrt(mse)      # root_mean_squared_error


print(f"MAE: {mae}\nMSE : {mse}\nRMSE : {rmse}\nAccuracy:{r2}")
```
