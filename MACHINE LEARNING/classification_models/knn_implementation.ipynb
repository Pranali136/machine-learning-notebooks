SPLITING THE DATA INTO TRAIN & TEST

```python
x=df.iloc[:,[2,3]] # Selects Age and EstimatedSalary as features.
y=df.iloc[:,-1]    #Selects Purchased column as target variable.
```

```python
#IMPORT MODELI to split data.
from sklearn.model_selection import  train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#Imports K-Nearest Neighbors classifier.
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Creates KNN model with K = 28 neighbors.
#K determines how many nearby points influence prediction.

knn=KNeighborsClassifier(n_neighbors=28)
```

```python
#Trains model using training data.
knn.fit(xtrain,ytrain)
#Predicts outcomes for test data.
ypred=knn.predict(xtest)
```

MODEL EVALUATION

```python
#Imports evaluation metrics.
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
ac = accuracy_score(ytest,ypred)   #Calculates overall accuracy.
cm= confusion_matrix(ytest,ypred)  #Shows correct vs incorrect predictions.
cr = classification_report(ytest,ypred)   #Displays precision, recall, f1-score.

print(f"Accuracy: {ac}\{cm}\{cr}")
```

##Train vs Test accuracy

```python
# Measures model performance.
# Used to check overfitting or underfitting.

train=knn.score(xtest,ytest)
test=knn.score(xtrain,ytrain)

print(f"train result {train}")
print(f"test result {test}")
```

```python
# Finding optimal K value
trainac = []
testac = []

#Tests K values from 1 to 30.
for i in range(1,31):
  #Trains model for each K.
  knn=KNeighborsClassifier(n_neighbors=i)
  knn.fit(xtrain,ytrain)

  #Calculates accuracy for each K.
  train=knn.score(xtrain,ytrain)
  test=knn.score(xtest,ytest)

  #Stores results.
  trainac.append(train)
  testac.append(test)


# print(f"{trainac}\n{testac}")

#Pairs train and test accuracy values.
list(zip(trainac,testac))
```

```python
# Visualizes accuracy vs K.
# Helps choose optimal K.
plt.plot(range(1,31),trainac,color="red")
plt.plot(range(1,31),testac,color="blue")
```

```python
#Defines a function to predict new users.
def purchased():
  #Takes user input.
  age=float(input('Enter a Age: '))
  salary=float(input("Enter a salary:"))
  #Formats input for prediction.
  newobs=[[age,salary]]
  #Predicts purchase outcome.
  ypred=knn.predict(newobs)[0]
  if (ypred==1):
    print("yes, he/she will purchased")
  else:
    print("No. he/she will not purchased")
  return ypred
```
