```python
x=df.iloc[:,:-1]# DataFrame
y=df.iloc[:,-1] #Series
```

```python
#Splits data into training (70%) and testing (30%)
#random_state ensures reproducibility

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
def mymodel(model):
  #model creation
  model.fit(xtrain,ytrain)  #Trains model on training data
  ypred = model.predict(xtest)  #Predicts output for unseen test data

  #check bias and variance

  train=model.score(xtrain,ytrain)  #Checks bias & variance
  test=model.score(xtest,ytest)
  print(f"training acccuracy:-{train}\ntesting accuracy:-{test}")

  #model evaluation
  print(classification_report(ytest,ypred))  #Class-wise performance analysis
  return model
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import  DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
KNN = mymodel(KNeighborsClassifier())
#high bias and low variance means model is underfitting.
```

```python
logreg=mymodel(LogisticRegression())
#low bias and high variance (overfiting)!apt.
```

```python
svm=mymodel(SVC())
#high bias and low variance means model is underfitting.
```

```python
DT=mymodel(DecisionTreeClassifier())
#low bias and high variance (overfiting)
```

Hyperparameter Tunning

```python
parameters = {
    "criterion":["gini","entropy"],
    "max_depth":list(range(1,20)),
    "min_samples_leaf":list(range(1,20))
}
```

GridSearchCV

GridSearchCV automatically finds the best hyperparameters for a model by trying all possible combinations and validating them.

```python
#CV = Cross-Validation
# Data is split into k parts (usually 5)
# Model is trained k-1 parts
# Tested on remaining part
# Repeated k times

from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(DecisionTreeClassifier(),parameters,verbose=2)  #Tries all parameter combinations
#verbose controls how much progress information is printed on the screen.
grid.fit(xtrain,ytrain)
```

```python
#Uses optimal Decision Tree
#Gives better generalization

dt=mymodel(grid.best_estimator_)
```
