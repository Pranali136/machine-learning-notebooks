```python
#IMPORT REQUIRED LIBRARIES


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

 BASIC LINE PLOT



```python
#Line plots are used to visualize trends over time or ordered data.
x = [1, 2, 3, 4, 5]                   # X-axis values
y = [10, 20, 30, 40, 50]              # Y-axis values

plt.plot(x, y)                        # Draws a line connecting x and y
plt.title("Basic Line Plot")          # Adds plot title
plt.xlabel("X Values")                # Labels X-axis
plt.ylabel("Y Values")                # Labels Y-axis
plt.show()                            # Displays the plot
```

CUSTOMIZING A LINE PLOT


```python
# Customization improves readability and presentation quality.
plt.figure(figsize=(6, 4))            # Controls figure size (width, height)
plt.plot(
    x, y,
    color="blue",                     # Changes line color
    linestyle="--",                   # Dashed line
    marker="o"                        # Adds markers on data points
)
plt.title("Customized Line Plot")
plt.show()
```

MULTIPLE LINE PLOTS


```python
#Used to compare multiple trends on the same graph
y1 = [10, 20, 30, 40, 50]
y2 = [15, 25, 35, 45, 60]

plt.plot(x, y1, label="2024 Sales")   # First line
plt.plot(x, y2, label="2025 Sales")   # Second line

plt.xlabel("Quarter")
plt.ylabel("Revenue")
plt.title("Sales Comparison")
plt.legend()                          # Explains line meaning
plt.grid(True)                        # Adds grid for readability
plt.show()
```

BAR CHART

```python
#Bar charts are used to compare values across categories.

categories = ["A", "B", "C", "D"]
values = [20, 35, 30, 15]

plt.bar(categories, values)           # Vertical bar chart
plt.title("Category Comparison")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

 HISTOGRAM



```python
#Histograms show the distribution of numerical data.
data = np.random.randint(1, 100, 50) # Random data for distribution

plt.hist(data, bins=10)               # bins define data grouping
plt.title("Data Distribution")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

 SCATTER PLOT

```python
# Scatter plots show relationships between two numerical variables.
x = [10, 20, 30, 40, 50]
y = [5, 15, 25, 35, 45]

plt.scatter(x, y, marker="*", s=100)  # s controls marker size
plt.title("Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()

```

PIE CHART

```python
# Pie charts show part-to-whole relationships.
labels = ["Python", "Java", "C++", "JavaScript"]
usage = [45, 25, 15, 15]

plt.pie(
    usage,
    labels=labels,
    autopct="%1.1f%%",                # Percentage formatting
    startangle=90
)
plt.title("Programming Language Usage")
plt.show()
```

SUBPLOTS (OBJECT-ORIENTED INTERFACE)

```python
# Subplots allow multiple plots in one figure.
# Object-Oriented approach is preferred in professional usage.
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(x, y1)                     # First subplot
ax[0].set_title("Line Plot")

ax[1].bar(categories, values)         # Second subplot
ax[1].set_title("Bar Plot")

plt.tight_layout()                    # Prevents overlap
plt.show()
```

AXIS CONTROL & TICKS

```python
# Axis limits and ticks help focus on specific data ranges.

plt.plot(x, y1)
plt.xlim(0, 6)                        # Limits X-axis
plt.ylim(0, 70)                       # Limits Y-axis
plt.xticks([1, 2, 3, 4, 5])            # Custom ticks
plt.title("Axis Control Example")
plt.show()
```

ANNOTATIONS

```python
# Annotations highlight important points in data.

plt.plot(x, y1)
plt.annotate(
    "Peak Value",
    xy=(5, 50),
    xytext=(3, 60),
    arrowprops=dict(arrowstyle="->")
)
plt.title("Annotation Example")
plt.show()
```

MATPLOTLIB WITH PANDAS

```python
# Real-world data is usually stored in DataFrames.

df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Sales": [12000, 15000, 25000, 45000]
})

plt.plot(df["Month"], df["Sales"])
plt.title("Monthly Sales (Pandas + Matplotlib)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

STYLES & THEMES

```python
# Styles improve plot aesthetics consistently.

plt.style.use("ggplot")
plt.plot(x, y1)
plt.title("Styled Plot (ggplot)")
plt.show()
```

 SAVING A PLOT

```python
# Used when exporting plots for reports or presentations.

plt.plot(x, y)
plt.title("Saved Plot Example")
plt.savefig("matplotlib_output.png")  # Saves plot as image
plt.show()
```
