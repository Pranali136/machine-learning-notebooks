```python
# Series using NumPy arrays
x=np.array([1,2,3])
y=np.array(["a","b","c"])
a=pd.Series(data=x,index=y)
a
```

DATAFRAME

```python
# Creating empty DataFrame
a=pd.DataFrame()
a
print( a)
```

```python
# DataFrame from list (each value becomes a row)
a=pd.DataFrame([1,2,3,4,5])
a
```

```python
# DataFrame from list variable
x=[1,2,3,4,5]
a=pd.DataFrame(data=x)
a
```

```python
# DataFrame from 2D list

d=[['i','ii','iii'],["alpha","beta","game"],["a","b","c"]]
a=pd.DataFrame(data=d,columns=["c1","c2","c3"])
a
```

```python
#Case Study(to arrange data column wise(vertical))
name=['pratik','purvesh','faizan','sandesh','madhu','vinayak']
py=[89,78,75,64,45,25]
ml=[89,57,65,78,25,35]

# Combine the lists as rows (each list becomes a row)
d = [name, py, ml]

# Create DataFrame with columns as the person names (1st list)
a = pd.DataFrame(data=d, index=["name","py","ml"])
a

#Vertical Formation
x={"name":name,"python":py,"ml":ml}    #The keys ("name", "python", "ml") represent the column names of your table.The values (name, py, ml) are lists that contain the actual data for each column
a=pd.DataFrame(x)     # The values (name, py, ml) are lists that contain the actual data for each column.Each key from x becomes a column header.
a
```
