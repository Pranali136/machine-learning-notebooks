RANGE-BASED ARRAY CREATION

```python
#functions of numpy
# arange() creates array with range of values
np.arange(1,7)
```

```python
x=np.arange(1,7)
```

```python
x.shape
```

```python
x.ndim
```

```python
# Reshaping into 2D
x=np.arange(1,7).reshape(2,3)
```

```python
x
```

```python
y=np.arange(7,13).reshape(2,3)
```

```python
y
```

```python
z
```

```python
z.shape
```

```python
z.size
```

```python
z.reshape(6,2)
```

```python
#hstack
np.hstack((x))
```

```python

#hstack → horizontal stacking (columns)
np.hstack((x,y))
```

```python
#vstack
np.vstack((x))
```

```python
# vstack → vertical stacking (rows)

np.vstack((x,y))
```
