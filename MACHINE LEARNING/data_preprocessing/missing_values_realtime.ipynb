IDENTIFY INVALID VALUES
* In real-world datasets, missing values are often represented
* using symbols like '?', 'NA', '-', etc.


```python

df['normalized-losses'].value_counts()
```

```python
df['horsepower'].value_counts()
```

```python
# REPLACE INVALID VALUES WITH NaN
df['normalized-losses'].replace("?",np.nan,inplace=True)
df['horsepower'].replace("?",np.nan,inplace=True)
```

```python
df['normalized-losses']
```

```python
df['horsepower'].value_counts()
```

```python
df
```

CHANGE DATA TYPE
* After replacing '?' with NaN, columns can be converted to float

```python

df['normalized-losses']=df['normalized-losses'].astype("float64")
```

```python
df['horsepower']=df['horsepower'].astype("float64")
```

```python
df.info()
```

```python
# CALCULATE MEAN VALUES
nlmean=df['normalized-losses'].mean()
hpmean=df['horsepower'].mean()
```

```python
nlmean
```

```python
hpmean
```

```python
# REPLACE MISSING VALUES USING MEAN
df['normalized-losses'].fillna(nlmean,inplace=True)
df['horsepower'].fillna(hpmean,inplace=True)
```
